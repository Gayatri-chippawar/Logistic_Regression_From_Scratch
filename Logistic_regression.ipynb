### Logistic Regression
$\sigma(x) = \frac{1}{1 + e^{-z}}$

z=w.X+b


### gradient Desccent


gradient descent is the optimization algorithm use for minimizing the loss function ,it is use for updating the paramenter of loss function 

$w = w - \alpha \, dw,\quad b = b - \alpha \, db$


Learning Rate
-------------

Learning rate is a tuning parameter in an optimization algorithm
that determines the step size at each iteration while moving
toward a minimum of a loss function.


Derivatives
-----------

$dw = \frac{1}{m}(\hat{Y} - Y)X,\quad db = \frac{1}{m}(\hat{Y} - Y)$




```python
#importing the dependenciers
import numpy as np 
```

**Scratch Model**

```python
class Logistic_Regression():
    def __init__(self,learn,iterate):
        self.learn=learn
        self.iterate=iterate

    def fit(self,X,Y):
        self.m,self.n=X.shape
        self.w=np.zeros(self.n)
        self.b=0
        self.X=X
        self.Y=Y
        for i in range(self.iterate):
            self.update_weights()

    def update_weights(self):
        z=self.X.dot(self.w)+self.b
        y_hat=1/(1+np.exp(-z))
        #defining the derivates 
        dw=(1/self.m)*np.dot(self.X.T,(y_hat-self.Y))
        db=(1/self.m)*np.sum(y_hat-self.Y)

        self.w=self.w-self.learn*dw
        self.b=self.b-self.learn*db

    def predict(self,X):
         z=X.dot(self.w)+self.b
         y_predict=1/(1+np.exp(-z))
         y_predict=np.where(y_predict>0.5,1,0)
         return y_predict
  
        
```
